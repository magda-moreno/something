## Prepare Data

➡️ Assignment Tasks
- Convert the Total_Amount into a number (not a string)
- Remove rows with empty (NULL) values
- Optional: Locate outliers in any of the variables.   Filter them out if they are mistakes; keep them if they are exceptional cases.

```python
df['total_amount'] = df['total_amount'].replace({'\$':'',',':''}, regex=True)
df['total_amount'] = pd.to_numeric(df['total_amount'])
df['total_amount']
```

```python
df['product_price'] = df['product_price'].replace({'\$':'',',':''}, regex=True)
df['product_price'] = pd.to_numeric(df['product_price'])
df['product_price']
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

## Summarize Data

➡️ Assignment Tasks
- Display the average price per month
- Display the total revenue for all months

```python
#average price
```

```python
df['order_month'].unique()
```

```python
df['order_month'] = pd.Categorical(df['order_month'], categories=['January ', 'February ', 'March ', 'April ', 'May ', 'June ', 'July ', 'August  ', 'September' , 'October ' , 'November ', 'December '], ordered=True)
df['order_month']
```

```python
month_grouped = df.groupby('order_month')
average_price = month_grouped['product_price'].mean().round(2)
average_price
```

```python
average_price.plot(kind= 'line',figsize=(10,5), xlabel='Month', ylabel='Product Price')
```

```python
average_price.plot(kind= 'bar', xlabel='Month', ylabel='Product Price')
```

```python
#total revenue
```

```python
tot_price = month_grouped['total_amount'].sum().round(2)
tot_price
```

```python
tot_price.plot(kind= 'line', figsize=(10,5), marker='*', xlabel='Month', ylabel='Product Price',color='red')
tot_price.plot(kind= 'bar', xlabel='Month', ylabel='Product Price')
```
